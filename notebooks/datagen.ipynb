```python
import matplotlib.pyplot as plt
import numpy as np
from shearnet.dataset import generate_dataset

# Set parameters
num_samples = 100  # Set this to 10000 or more for thousands of images
psf_fwhm = 1.0      # PSF full width at half maximum

# Generate dataset for the ideal PSF model
images_ideal, labels_ideal = generate_dataset(num_samples, psf_fwhm, exp='ideal')

# Generate dataset for the superbit PSF model
images_superbit, labels_superbit = generate_dataset(num_samples, psf_fwhm, exp='superbit')
```

```python
# Display the shapes of the generated datasets
print("Ideal PSF dataset:")
print("Images shape:", images_ideal.shape)
print("Labels shape:", labels_ideal.shape)

print("\nSuperbit PSF dataset:")
print("Images shape:", images_superbit.shape)
print("Labels shape:", labels_superbit.shape)

# Plot a few example images from each dataset
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Ideal PSF examples
axes[0, 0].imshow(images_ideal[0], cmap='gray')
axes[0, 0].set_title(f'Ideal PSF\nShear: {labels_ideal[0]}')
axes[0, 0].axis('off')

axes[0, 1].imshow(images_ideal[1], cmap='gray')
axes[0, 1].set_title(f'Ideal PSF\nShear: {labels_ideal[1]}')
axes[0, 1].axis('off')

# Superbit PSF examples
axes[1, 0].imshow(images_superbit[25], cmap='gray')
axes[1, 0].set_title(f'Superbit PSF\nShear: {labels_superbit[0]}')
axes[1, 0].axis('off')

axes[1, 1].imshow(images_superbit[90], cmap='gray')
axes[1, 1].set_title(f'Superbit PSF\nShear: {labels_superbit[1]}')
axes[1, 1].axis('off')

plt.tight_layout()
plt.show()

# Print the first 100 superbit images without labels in a grid
fig, axes = plt.subplots(10, 10, figsize=(10, 10))
axes = axes.ravel()

for i in range(99):
    axes[i].imshow(images_superbit[i], cmap='gray')
    axes[i].axis('off')

plt.tight_layout()
plt.show()

```
